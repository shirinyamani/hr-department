```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/data.csv')
```

```python
#mean of the age
df['Age'].mean()
```

```python
df.describe()
```

```python
#replace the Yes/No in attritittion, overtime, over18 colums
mapping = {'Yes':1, 'Y':1, 'No':0, 'N':0}
df = df.applymap(lambda x:mapping.get(x,x))

#same as if you want to do it manually for each of the cols
df['Over18']=df['Over18'].apply(lambda x: 1 if x=='Yes' or 'Y' else 0)
```

```python
#check for null values
df.isnull().sum()
df.isna().any().any()
```

```python
#count the number of the 1s in the colums
(df['EmployeeCount']== 1).sum() #col wise
df[df['Attrition']==1].shape[0] #row wise
set(df['Education']) #see different values 
```

```python
#remove the non-informative cols
df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
df.shape #as we drop 4 non-uselful cols
```

```python
#see the ppl who left the company
left_df = df[df['Attrition'] == 1]
stayed_df = df[df['Attrition'] == 0]
```

```python
print(f'left ppl:{len(left_df)}, stayed ppl:{len(stayed_df)}')
```

```python
left_df.describe()
```

```python
stayed_df.describe()
```

```python
plt.figure(figsize=[25, 12])
sns.countplot(x = 'Age', hue = 'Attrition', data = df)
```

```python
plt.figure(figsize=(12,7))

sns.kdeplot(left_df['DistanceFromHome'], label = 'Employees who left', shade = True, color = 'r')
sns.kdeplot(stayed_df['DistanceFromHome'], label = 'Employees who Stayed', shade = True, color = 'b')

plt.xlabel('Distance From Home')
```
